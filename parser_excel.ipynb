```python
import pandas as pd
```

```python
stat = pd.read_excel('./statistics.xlsx')
stat.head()
```

```python
#rating=pd.read_table('./rating.txt',sep='\n')
#rating

with open("./rating.txt", "r",encoding="utf-8") as f:
    text = f.read()
    print(text)
```

```python
lst = text.split("\n")
for i in lst:
    if i=='':
        lst.remove('')
lst.remove('')
len(lst)
```

```python

for i in range(len(lst)):
    l=lst[i].split()
    for j in range(len(l)):
        if l[j]=='Number:':
            l=l[j:]
            break
    lst[i]=" ".join(l)
    
num=[]    
for i in range(len(lst)):
    l=lst[i].split()
    num.append(l[1])
    
for i in range(len(lst)):
    l=lst[i].split()
    l=l[2:]
    lst[i]=" ".join(l)
```

```python
stat.insert(2,"text", lst, True)
stat.insert(2,"Number", num, True)
stat
```

```python
writer = pd.ExcelWriter('stat_1.xlsx')
stat.to_excel(writer)
writer.save()
```
